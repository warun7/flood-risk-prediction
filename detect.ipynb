```python
#loading datasets
df_factors = pd.read_csv('./datasets/floods_factors/flood.csv')
df_historical = pd.read_csv('./datasets/floods_inventory/India_Floods_Inventory.csv')
df_rainfall = pd.read_csv('./datasets/rainfall_data/rainfall.csv')

#datasets info
print(df_factors.info())
print(df_historical.info())
print(df_rainfall.info())
```

```python
df_historical.isnull().sum()
#drop missing latitude and longitude values
df_historical = df_historical.dropna(subset=['Latitude', 'Longitude'])

df_historical['Start Date'] = pd.to_datetime(df_historical['Start Date'], errors='coerce')
df_historical['End Date'] = pd.to_datetime(df_historical['End Date'], errors='coerce')

start_limit = pd.Timestamp("2015-06-27")
end_limit = pd.Timestamp("2025-03-05")

df_historical = df_historical[
    (df_historical['Start Date'] >= start_limit) & 
    (df_historical['End Date'] <= end_limit)
]

df_extraction_data = df_historical[['Latitude', 'Longitude', 'Start Date', 'End Date']]

df_extraction_data.to_csv(r'./datasets/floods_inventory/info.csv', index=False, header=True)
```
